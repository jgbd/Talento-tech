# Import Librery Pandas

```python
import pandas as pd
import random
```

## Defnined Variables and created DataFrame

```python
name_list = ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack']
weight_list = [55, 85, 75, 95, 65, 80, 70, 60, 90, 100]
height_list = [1.65, 1.80, 1.75, 1.90, 1.60, 1.85, 1.70, 1.55, 1.95, 2.00]

data_dict = {
    'Name': name_list,
    'Weight': weight_list,
    'Height': height_list
}

df = pd.DataFrame(data_dict)
```

## Print DataFrame

```python
df
```

## Add records automatic

```python
def add_records(df, num_records):
    for i in range(num_records):
        weight = random.randint(50, 100)
        height = random.uniform(1.0, 2.0)
        new_record = pd.DataFrame({'Name': [f'Person_{i+1}'], 'Weight': [weight], 'Height': [height]})
        df = pd.concat([df, new_record], ignore_index=True)
    return df
```

## execute add records

```python
df = add_records(df, 1000)
```

## Define Function classify BMI

```python
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'
```

## Calculate BMI

```python
df['bmi'] = df['Weight'] / (df['Height'] ** 2)
```

## Apply function and add new column category

```python
df['BMI Category'] = df['bmi'].apply(classify_bmi)
```

Visualizar todo el Dataframe

## metricas de regresion y de clasificacion

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score, classification_report

true_bmi_category = df['BMI Category']
predicted_bmi_category = df['BMI Category']

accuracy = accuracy_score(true_bmi_category, predicted_bmi_category)
report = classification_report(true_bmi_category, predicted_bmi_category)

print("Classification Metrics:")
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)
```

## ejemplo de matrix de correlacion y otro de confusion

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

correlation_matrix = df[['Weight', 'Height', 'bmi']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Weight, Height, and BMI')
plt.show()

conf_matrix = confusion_matrix(df['BMI Category'], df['BMI Category'])
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=df['BMI Category'].unique(), yticklabels=df['BMI Category'].unique())
plt.xlabel('Predicted BMI Category')
plt.ylabel('True BMI Category')
plt.title('Confusion Matrix')
plt.show()
```
